# Exploratory Data Analysis (EDA) of Marvel and DC Characters Data 

```python
# Loading the required libraries
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
marvel=pd.read_csv('../input/marvelanddccharactersdataset/charcters_stats.csv')
```

```python
marvel.head()
```

```python
marvel.shape
```

```python
marvel['Alignment'].value_counts()
```

```python
marvel[marvel['Alignment']=='good']
```

```python
good=marvel[marvel['Alignment']=="good"]
```

```python
good.head()
```

```python
good.sort_values(by=['Speed'],ascending=False)
```

```python
good.sort_values(by=['Power'],ascending=False)
```

```python
good_max_power_full=good[good['Power']==100]
```

```python
good_max_power_full.shape
```

```python
good_max_power_full
```

```python
good_max_power=good.sort_values(by=['Total'],ascending=False)
```

```python
good_max_power.head()
```

```python
plt.figure(figsize=(8,5))
plt.bar(list(good_max_power['Name'])[0:5],list(good_max_power['Total'])[0:5],color="g")
plt.show()
```

```python
bad=marvel[marvel['Alignment']=="bad"]
```

```python
bad.head()
```

```python
bad.sort_values(by=['Speed'],ascending=False).head()
```

```python
bad.sort_values(by=['Intelligence'],ascending=False).head()
```

```python
bad.sort_values(by=['Total'],ascending=False).head()
```

```python
bad_max_power=bad.sort_values(by=['Total'],ascending=False)
```

```python
plt.figure(figsize=(10,5))
plt.bar(list(bad_max_power['Name'])[0:5],list(bad_max_power['Total'])[0:5],color=["red","blue","green","brown","black"])
plt.show()
```

```python
# Making a histogram for speed of good superheroes 

plt.figure(figsize=(10,6))
plt.hist(good['Speed'])
plt.title("Distribution of speeds")
plt.xlabel("Speed")
plt.show()
```

```python
# Making a histogram for Combat of good supervillains 

plt.figure(figsize=(10,6))
plt.hist(good['Combat'])
plt.title("Distribution of Combat")
plt.xlabel("Combat")
plt.show()
```
